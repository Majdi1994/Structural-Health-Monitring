## Data Description  

CWRU a single point failure dataset that constructed from 3 main faults each fault consist of 3 damage sizes. The total number of fault classes will be 10 classes (i.e. 1 normal class and 9 healthy classes), here is details about the data acquisition : 
- Data is collected at sampling frequency 12,000 samples/second. 
- Each data file lasts for 10 seconds. Thus, it contains 120000 datapoint. 
- Data is collected under 4 different loading conditions (i.e. 0,1,2, and 3 HP)
- Each working condition has 10 classes, and each class has has time series with length 120000 data points

## Functions

```python
def loadMetadata():
    f_in = os.path.join(os.getcwd(),'CWRU-raws','CWRU-metadata.txt')
    dfMetadata = pd.read_csv(f_in,sep='\t')
    fault_code = dfMetadata['Inner_race'].map(str) + \
        dfMetadata['Ball'].map(str) + \
        dfMetadata['Outer_race_centered'].map(str) + \
        dfMetadata['Outer_race_orthogonal'].map(str) + \
        dfMetadata['Outer_race_opposite'].map(str)
    dfMetadata['Fault_type'] = \
        fault_code.map({'00000':'normal','10000':'inner','01000':'ball','00100':'outer-cen',
        '00010':'outer-ort','00001':'outer-opp'})
    return dfMetadata

def loadTimeseriesSensorData(defect,sampling_rate,fault_diameter,motor_load,fault_type,end):
    
    # fault_type must be ether 'inner', 'ball', 'outer-cen', outer-ort',
    # or outer-opp; end is either DE (for data generated at drive-end) or 
    # FE (for data generated at fan-end)
    f_in = dfMetadata[(dfMetadata.Defect==defect) &\
        (dfMetadata.Sampling_rate==sampling_rate) &\
        (dfMetadata.Fault_diameter.round(decimals=3)==fault_diameter) &\
        (dfMetadata.Motor_load==motor_load) &\
        (dfMetadata.Fault_type==fault_type)].iloc[0].File
    bearing_data = scipy.io.loadmat(os.path.join(os.getcwd(),'CWRU-raw',f_in))
    
    if len(f_in)==6: f_in = '0'+f_in
    if end=='DE':
        return bearing_data['X'+f_in.strip('.mat')+'_DE_time']
    else:
        return bearing_data['X'+f_in.strip('.mat')+'_FE_time']

def visualizeTimeseriesSensorData(timeseries,window_start=0,window_end=20000):
    plt.figure(0,figsize=(100,20))        
    plt.plot (timeseries[window_start:window_end],'red')
    plt.show()
```

## Examples for functions

```python
# Load metadata of CWRU bearing data
dfMetadata = loadMetadata()

## Examples

# Load specific timeseries CWRU bearing data
defect = 'normal'
sampling_rate = '12k'
fault_diameter = 0 
motor_load = 0
fault_type = 'normal'
end = 'DE'
timeseries = loadTimeseriesSensorData(defect,sampling_rate,fault_diameter,motor_load,fault_type,end)
```

## Loading timeseries data of interest

```python
## We load timeseries data collected at drive-end for:
##   i)   healthy data under motor load of 0 HP (normal_0), 1 HP (normal_1), 2 HP (normal_2) and 3 HP (normal_3);
##   ii)  faulty data under each combination of the following conditions:
##          a) inner fault, outer fault or ball-centered fault, 
##          b) fault diameter of 0.007 inches, 0.014 inches or 0.021 inches,
##          c) motor load of 0 HP (normal_0), 1 HP (normal_1), 2 HP (normal_2) or 3 HP (normal_3)

# Load metadata of CWRU bearing data
dfMetadata = loadMetadata()

# normal

normal_0 = loadTimeseriesSensorData('normal','12k',0,0,'normal','DE')
normal_1 = loadTimeseriesSensorData('normal','12k',0,1,'normal','DE')
normal_2 = loadTimeseriesSensorData('normal','12k',0,2,'normal','DE')
normal_3 = loadTimeseriesSensorData('normal','12k',0,3,'normal','DE')

# inner fault

ir007_0 = loadTimeseriesSensorData('drive-end','12k',0.007,0,'inner','DE')
ir007_1 = loadTimeseriesSensorData('drive-end','12k',0.007,1,'inner','DE')
ir007_2 = loadTimeseriesSensorData('drive-end','12k',0.007,2,'inner','DE')
ir007_3 = loadTimeseriesSensorData('drive-end','12k',0.007,3,'inner','DE')

ir014_0 = loadTimeseriesSensorData('drive-end','12k',0.014,0,'inner','DE')
ir014_1 = loadTimeseriesSensorData('drive-end','12k',0.014,1,'inner','DE')
ir014_2 = loadTimeseriesSensorData('drive-end','12k',0.014,2,'inner','DE')
ir014_3 = loadTimeseriesSensorData('drive-end','12k',0.014,3,'inner','DE')

ir021_0 = loadTimeseriesSensorData('drive-end','12k',0.021,0,'inner','DE')
ir021_1 = loadTimeseriesSensorData('drive-end','12k',0.021,1,'inner','DE')
ir021_2 = loadTimeseriesSensorData('drive-end','12k',0.021,2,'inner','DE')
ir021_3 = loadTimeseriesSensorData('drive-end','12k',0.021,3,'inner','DE')

# ball fault

b007_0 = loadTimeseriesSensorData('drive-end','12k',0.007,0,'ball','DE')
b007_1 = loadTimeseriesSensorData('drive-end','12k',0.007,1,'ball','DE')
b007_2 = loadTimeseriesSensorData('drive-end','12k',0.007,2,'ball','DE')
b007_3 = loadTimeseriesSensorData('drive-end','12k',0.007,3,'ball','DE')

b014_0 = loadTimeseriesSensorData('drive-end','12k',0.014,0,'ball','DE')
b014_1 = loadTimeseriesSensorData('drive-end','12k',0.014,1,'ball','DE')
b014_2 = loadTimeseriesSensorData('drive-end','12k',0.014,2,'ball','DE')
b014_3 = loadTimeseriesSensorData('drive-end','12k',0.014,3,'ball','DE')

b021_0 = loadTimeseriesSensorData('drive-end','12k',0.021,0,'ball','DE')
b021_1 = loadTimeseriesSensorData('drive-end','12k',0.021,1,'ball','DE')
b021_2 = loadTimeseriesSensorData('drive-end','12k',0.021,2,'ball','DE')
b021_3 = loadTimeseriesSensorData('drive-end','12k',0.021,3,'ball','DE')

# outer fault, centered

or007at6_0 = loadTimeseriesSensorData('drive-end','12k',0.007,0,'outer-cen','DE')
or007at6_1 = loadTimeseriesSensorData('drive-end','12k',0.007,1,'outer-cen','DE')
or007at6_2 = loadTimeseriesSensorData('drive-end','12k',0.007,2,'outer-cen','DE')
or007at6_3 = loadTimeseriesSensorData('drive-end','12k',0.007,3,'outer-cen','DE')

or014at6_0 = loadTimeseriesSensorData('drive-end','12k',0.014,0,'outer-cen','DE')
or014at6_1 = loadTimeseriesSensorData('drive-end','12k',0.014,1,'outer-cen','DE')
or014at6_2 = loadTimeseriesSensorData('drive-end','12k',0.014,2,'outer-cen','DE')
or014at6_3 = loadTimeseriesSensorData('drive-end','12k',0.014,3,'outer-cen','DE')

or021at6_0 = loadTimeseriesSensorData('drive-end','12k',0.021,0,'outer-cen','DE')
or021at6_1 = loadTimeseriesSensorData('drive-end','12k',0.021,1,'outer-cen','DE')
or021at6_2 = loadTimeseriesSensorData('drive-end','12k',0.021,2,'outer-cen','DE')
or021at6_3 = loadTimeseriesSensorData('drive-end','12k',0.021,3,'outer-cen','DE')
```

```python
## Constructing datasets for different motor loads (working conditions).
##
## Here we will take concatenate all classes corresponding to each working condition. 
## Eventually, we will have 4 working conditions each with dimension 120000 x 10, 
## with 120000 data length and 10 classes. 
## Here we load matlab matrices for each class type, the length of each file is about 120000 datapoint 
## ( i.e. 12000 sample/seconds lasts for 10 seconds).


# working condition_0 with no load
num_points=120000
motor_load_0hp = np.concatenate((normal_0[0:num_points],
    ir007_0[0:num_points],b007_0[0:num_points],or007at6_0[0:num_points],
    ir014_0[0:num_points],b014_0[0:num_points],or014at6_0[0:num_points],
    ir021_0[0:num_points],b021_0[0:num_points],or021at6_0[0:num_points]),
    axis=1)

# working condition_1 with load 1 hp
motor_load_1hp = np.concatenate((normal_1[0:num_points],
    ir007_1[0:num_points],b007_1[0:num_points],or007at6_1[0:num_points],
    ir014_1[0:num_points],b014_1[0:num_points],or014at6_1[0:num_points],
    ir021_1[0:num_points],b021_1[0:num_points],or021at6_1[0:num_points]),
    axis=1)

# working condition_2 with load 2 hp
motor_load_2hp = np.concatenate((normal_2[0:num_points],
    ir007_2[0:num_points],b007_2[0:num_points],or007at6_2[0:num_points],
    ir014_2[0:num_points],b014_2[0:num_points],or014at6_2[0:num_points],
    ir021_2[0:num_points],b021_2[0:num_points],or021at6_2[0:num_points]),
    axis=1)

# working condition_3 with load 3 hp
motor_load_3hp = np.concatenate((normal_3[0:num_points],
    ir007_3[0:num_points],b007_3[0:num_points],or007at6_3[0:num_points],
    ir014_3[0:num_points],b014_3[0:num_points],or014at6_3[0:num_points],
    ir021_3[0:num_points],b021_3[0:num_points],or021at6_3[0:num_points]),
    axis=1)
```

### Saving datasets under different motor loads (working conditions) as torch tensor

```python
motor_load_0hp = torch.from_numpy(motor_load_0hp)
motor_load_1hp = torch.from_numpy(motor_load_1hp)
motor_load_2hp = torch.from_numpy(motor_load_2hp)
motor_load_3hp = torch.from_numpy(motor_load_3hp)
```

## Data Augmentation using Sliding Window
In order to develop sample with suitable size for training the model, sliding window with 4096 length (i.e. sample length): 
- Without window overlapping we will have about _30 samples_ from _each class_ to train the model, which is not enough.
- To have enoguh training samples we incorporated sliding window with overlapping, _shifting size= 290_. Thus, we got about _400 sample_ for _each class_ and total_4000 sample_ for each _working condition_. The following equations shows how we calcualte overlapping:

\begin{align}
n_{class} & =(\frac{total\_length - sample\_length}{shifting\_size}) \\ 
N_{wk} &= n_{class}* 10 
\end{align}

Where $n_{class}$ is the number of samples per each class and $N_{wk}$ is total number of samples for each working condition. 
![alt text](window.png "Title")

```python
# training data extraction 
# variables initialization
i = 0
len_data = 120000
# total number of samples for all classes 
# num_sample_per_class=num_sample/10
# number with augmented data
num_samples = 4030
# number with no overlap data
#num_samples=1000
#sample_len=1200
sample_len = 4096

# without overlapping
stride = sample_len

# with overlap
overlap = 1
if overlap: stride = 290
    
# initialize train data and labels
work_cond_0hp = torch.zeros(num_samples,sample_len)
work_cond_1hp = torch.zeros(num_samples,sample_len)
work_cond_2hp = torch.zeros(num_samples,sample_len)
work_cond_3hp = torch.zeros(num_samples,sample_len)
data_labels = torch.LongTensor(num_samples)

for class_select in range(10):
    for count in range(0,len_data-4096,stride):
        work_cond_0hp[i] = motor_load_0hp[count:count+sample_len,class_select]
        work_cond_1hp[i] = motor_load_1hp[count:count+sample_len,class_select]
        work_cond_2hp[i] = motor_load_2hp[count:count+sample_len,class_select]
        work_cond_3hp[i] = motor_load_3hp[count:count+sample_len,class_select]
        data_labels[i] = class_select
        i += 1

### Cutting down to 4000 samples
work_cond_0hp = work_cond_0hp[0:4000]
work_cond_1hp = work_cond_1hp[0:4000]
work_cond_2hp = work_cond_2hp[0:4000]
work_cond_3hp = work_cond_3hp[0:4000]
data_labels = data_labels[0:4000]
```

## Splitting data into training and testing 

```python
train_size = int(0.5 * len(work_cond_0hp))
test_size = len(work_cond_0hp) - train_size
train_idx, test_idx = torch.utils.data.random_split(np.arange(len(work_cond_0hp)), [train_size, test_size])

# 0 HP working condition
train_0hp_data = work_cond_0hp[train_idx]
train_0hp_labels = data_labels[train_idx]
test_0hp_data = work_cond_0hp[test_idx]
test_0hp_labels = data_labels[test_idx]

# 1 HP working condition  
train_1hp_data = work_cond_1hp[train_idx]
train_1hp_labels = data_labels[train_idx]
test_1hp_data = work_cond_1hp[test_idx]
test_1hp_labels = data_labels[test_idx]

# 2 HP working condition
train_2hp_data = work_cond_2hp[train_idx]
train_2hp_labels = data_labels[train_idx]
test_2hp_data = work_cond_2hp[test_idx]
test_2hp_labels = data_labels[test_idx]

# 3 HP working condition 
train_3hp_data = work_cond_3hp[train_idx]
train_3hp_labels = data_labels[train_idx]
test_3hp_data = work_cond_3hp[test_idx]
test_3hp_labels = data_labels[test_idx]
```
